```python
import pandas as pd
import json
```

```python
with open('../output/characters_raw.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
characters_df = pd.DataFrame(data['characters'])
```

```python
characters_df.head(10)
```

```python
# Need to remove the wiki reference tags [1][2] and such from the data
for col in characters_df:
    characters_df[col]=characters_df[col].str.replace(r'\[\d+\]', '', regex=True)
```

```python
characters_df.head(10)
```

```python
characters_df.to_json('../output/full-character-data-processed.json', orient='records', force_ascii=False,indent=2)
```
